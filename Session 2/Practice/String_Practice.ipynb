Challenge Question

Write a program that:

Asks the user to enter a sentence

Prints:

The number of characters

The number of words

The sentence in uppercase

The sentence reversed

```python
Sentence=input("Enter Sentence : ")
```

```python
print("Length of Sentence is : ",len(Sentence))
```

```python
words=Sentence.split()
print("Number of Words is : ",len(words))
```

```python
print("Sentence UpperCased : ",Sentence.upper())
print("Sentence Titled : ",Sentence.title())
```

```python
print("Reversed Words : ", words[::-1])
print("Reversed Sentence : ", " ".join(words[::-1]))
print("Reversed Sentence and Characters : ", Sentence[::-1])
```
