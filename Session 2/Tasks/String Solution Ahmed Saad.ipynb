String Methods
Write a program that:
Asks the user to enter a text
Prints the text in uppercase
Prints the text in lowercase
Prints the number of characters in the text

```python
str_UPPER_And_LOWER_Case=input("Enter String ")
# Input : ahmed saad 
```

```python
# Capitalized All String
print(str_UPPER_And_LOWER_Case.upper())
# Capitalized First Character
print(str_UPPER_And_LOWER_Case.capitalize())
```

```python
# lower All String
print(str_UPPER_And_LOWER_Case.lower())
# lower First Character
print(str_UPPER_And_LOWER_Case[0].lower() + str_UPPER_And_LOWER_Case[1:])
# Input : AHMED SAAD 
```
