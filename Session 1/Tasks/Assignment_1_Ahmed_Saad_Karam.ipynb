## Create a Python program that:

### 1- Asks the user to enter two numbers.

### 2-Stores them as variables (make sure to convert them to integers).

### 3-Performs and prints the results of the following operations:

* ## Addition

* ## Subtraction

* ## Multiplication

* ## Division

* ## Power (num1 ** num2)

* ## Modulus (num1 % num2)

* ## Floor division (num1 // num2)






```python
first_number=int(input("Enter First Number : "))
second_number=int(input("Enter Second Number : "))
```

```python
addition=first_number+second_number
subtraction=first_number-second_number
multiplication=first_number*second_number
power=first_number**second_number
modulus=first_number%second_number
floor_division=first_number//second_number
```

```python
from __future__ import division
if(second_number==0):
    print("Math Error")
else:
  division=first_number/second_number
```

```python
print("First Number Added by Second Number Equal : ", addition)
print("First Number Subtracted by Second Number Equal : ", subtraction)
print("First Number Multiplied by Second Number Equal : ", multiplication)
print("First Number Power by Second Number Equal : ", power)
print("First Number Modulus by Second Number Equal : ", modulus)
print("First Number Floor Division by Second Number Equal : ", floor_division)
print("First Number Division by Second Number Equal : ", division)
```

# Question 3
# Write a Python program that does the following:

# Asks the user to enter the length and width of a rectangle (Float).

# Calculates the area and perimeter of the rectangle(Float).

# Prints the result neatly, for example:
# Enter length: 5
# Enter width: 3
# Area: 15.0
# Perimeter: 16.0

```python
lenghth=float(input("Enter Lenghth : "))
width=float(input("Enter Width : "))
```

```python
area = lenghth * width
perimeter = 2 * (lenghth + width)
```

```python
print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
```
